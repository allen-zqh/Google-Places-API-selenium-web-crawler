```python
import requests
import json
from tqdm import tqdm
import pandas as pd

API_KEY = ''
df = pd.read_excel('/Users/zqh/Documents/RIETI_new/RIETI_AI_Survey/todo_test.xlsx')

for ind, row in tqdm(df.iterrows()):
    name = df.loc[ind, 'name']
    geocoding = '%40' + str(df.loc[ind, 'fY']) + '%2C' + str(df.loc[ind, 'fX'])    
    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?fields=formatted_address%2Cname%2Cgeometry%2Cplace_id&input=' + name + '&inputtype=textquery&locationbias=circle%3A2000' + geocoding + '&key=' + API_KEY

    r1 = requests.get(url)
    r1_json = json.loads(r1.text)
    if r1_json.get('candidates'):
        place_id = r1_json['candidates'][0].get('place_id')
        google_name = r1_json['candidates'][0].get('name')
        google_address = r1_json['candidates'][0].get('formatted_address')
        google_geo_lng = r1_json['candidates'][0].get('geometry').get('location').get('lng')
        google_geo_lat = r1_json['candidates'][0].get('geometry').get('location').get('lat')
        
        place_url = 'https://maps.googleapis.com/maps/api/place/details/json?place_id=' + place_id + '&fields=website&key=' + API_KEY
        r2 = requests.get(place_url)
        r2_json = json.loads(r2.text)
        try:
            firm_web = r2_json['result']['website']
        except:
            firm_web = None
        
        df.loc[ind, 'google_name'] = google_name
        df.loc[ind, 'google_placeId'] = place_id
        df.loc[ind, 'google_address'] = google_address
        df.loc[ind, 'google_geo_lng'] = google_geo_lng
        df.loc[ind, 'google_geo_lat'] = google_geo_lat
        df.loc[ind, 'website'] = firm_web
```
